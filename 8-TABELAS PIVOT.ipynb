# **TABELAS PIVOT**

```python
# Tabelas pivot são uteis para agrupamento de dados.
#Leitura do dataset.
import pandas as pd
arquivo = 'C:/Users/William Cararo/Videos/Curso - Data Science/Módulo 5 - python para analise de dados/pandas/materiais-de-apoio-pandas/materiais-de-apoio-pandas/kc_house_data.csv'

dataset = pd.read_csv(arquivo, sep=',')
```

```python
#Margins coloca o total na tabela.
dataset.pivot_table('id', index=['waterfront','floors'],aggfunc='count', margins=True)
```

```python
# Crosstab ou tabulação cruzada. Cruza valores das variaveis
pd.crosstab(dataset['bedrooms'],dataset['condition'])
```

```python
# Plota uma tabulação cruzada em um gráfico.
#necessário o import do matplot notebook para redimensionamento e etc...
import matplotlib.pyplot as plt
%matplotlib notebook 
table = pd.crosstab(dataset['bedrooms'],dataset['condition'])
table.plot(kind='bar', width=1.0, color=['r','b','y'], title='condition by bedrooms', grid=False)
```
